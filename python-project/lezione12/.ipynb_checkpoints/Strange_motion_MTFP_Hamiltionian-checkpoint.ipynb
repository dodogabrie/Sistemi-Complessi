```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from scipy.stats import kde
```

```python
def get_param(): # Scelgo di definirli in una funzione e poi chiamarla successivamente, 
                 # non voglio creare var. globali.
    """Parametri del problema""" 
    dt = 0.01
    A = 0.3
    w = 1
    n = 1000
    m = 100
    return dt, A, w, n, m
```

```python
def f(x):
    """Potenziale del moto"""
    return x - x**3

def step(x1, x, p2):
    """Step per il camminatore vero"""
    p = get_param()
    dt, A = p[:2]
    return (f(x) + A * np.sin(x1) + p2) * dt

def step_x1():
    """Step per il camminatore ausiliario"""
    p = get_param()
    dt, w = p[0], p[2]
    return w * dt

def step_p1(x1, p2):
    """Step per l'impulso sul camminatore ausiliario"""
    p = get_param()
    dt, A = p[:2]
    return (- A * np.cos(x1) * p2) * dt

def step_p2(x, p2):
    """Step per l'impulso sul camminatore vero"""
    p = get_param()
    dt = p[0]
    return ( ( -1 + 3 * x**2 ) * p2 ) * dt
```

```python
def init_cond(array_in = None, init=False, mode = 'zeros'):
    """Funzione per generare delle condizioni iniziali su m cammini a n step"""
    par = get_param()
    n, m = par[3], par[4]
    if init: # Se voglio inizializzare un array personalizzato entro qui
        x_init = array_in                  # Condizioni iniziali
        x_aux = np.ones((n-1 , m))         # Resto della matrice
        fin = np.vstack((x_init,x_aux))    # Unisco i vettori
        return fin
    else: # Se voglio un array standard entro qui
        if mode == 'zeros': # Array di zeri
            return np.zeros((n , m))
        if mode == 'ones': # array di 1 (sfruttabile per essere moltiplicato per una costante...)
            return np.ones((n , m))
```

Il fatto che i cammini scappino rende il codice complicato da fare in modo vettoriale (overflow), quindi ho dovuto scrivere i cicli in modo "standard"... Questo significa in python andare molto piano. Il codice potrebbe essere reso vettoriale facendo uso (forse) della funzione di Heavyside per i più coraggiosi, l'alternativa è scriverlo in Fortran! :(

```python
def motion(x, x1, p1, p2, n, m):
    '''Gestione del movimento del camminatore'''
    aux = x # Metto l'array in ingresso in una variabilie ausiliaria per non perderlo
    p = get_param()
    dt, A, w = p[:3]
    print('start cycling')
    s = []
    if m > 1: # Se abbiamo più cammini devo ciclare su due indici (in python questo rende il codice estremamente lento)
        for j in range(m):
            for i in range(n-1):
                if np.abs(x[i,j]) <= 1.5: # Sopra a questo valore stabilizzo i camminatori: scapperanno e 
                                          # non torneranno indietro.
                    x1[i+1,j] = x1[i,j] + step_x1()
                    x[i+1,j]  = x[i,j]  + step( x1[i,j], x[i,j], p2[i,j] ) 
                    p1[i+1,j] = p1[i,j] + step_p1( x1[i,j], p2[i,j] )
                    p2[i+1,j] = p2[i,j] + step_p2( x[i,j], p2[i,j] )
                else:
                    s.append(i)
                    x[i+1,j] = x[i,j] # Lascio i camminatori a 1.5
    else:
        for i in range(n-1): # se in ingresso c'è un vettore allora evito errore sull'uso degli indici.
            if np.abs(x[i]) <= 1.5:
                x1[i+1] = x1[i] + step_x1()
                x[i+1]  = x[i]  + step( x1[i], x[i], p2[i] ) 
                p1[i+1] = p1[i] + step_p1( x1[i], p2[i] )
                p2[i+1] = p2[i] + step_p2( x[i], p2[i] )
            else:
                s.append(i)
                x[i+1] = x[i]
    print('End cycling')
    return x, p2, s # Return del vettore finale x e di s: un array contenente gli indici fuggiti
```

```python
def main():
    """Creo i plot e inizializzo i vettori"""
    par = get_param()
    n, m = par[3], par[4] # prendo i parametri del problema
    dt, w = par[0], par[2]
    #plt.figure(figsize=(20,10))
    
    # Setting assi e limiti del moto
    T = 2*np.pi/w # Estraggo il periodo della perturbazione oscillante
    n1 = 2*n # Limite ausiliario al plotting 
    t = np.arange(-T,n1*dt-T,dt) + 2.5 # Array per le x
    x_b = -1.045*np.ones(n1)  # Il valore -1.045 pone il camminatore inizialmente sulla cresta della perturbazione
    x1_b = np.zeros(n1)  # Metto questi nulli così il camminatore rimane lì
    p1_b = np.zeros(n1)
    p2_b = np.zeros(n1)
    x_b, p_b, s = motion(x_b, x1_b, p1_b, p2_b, n1, 1) # Evolvo 
    plt.plot(t, x_b, color = 'black', linestyle='--') # Plotto 
    
    x_b, p_b, s = motion(.945*np.ones(n1), x1_b, p1_b, p2_b, n1, 1) # creo moto anche intorno a -1 (altro equilibrio)
    plt.plot(t, x_b-1, color='black', linestyle='--') # traslo di -1 per portarlo all'estremo nullo
    
    t_short = np.arange(-T, n1*dt - T,dt) + 2.5  # Array che parte più indietro per far sembrare i camminatori 
                                                 # Inizialmente sulla traiettoria (linea blu)
    x_b, p_b, s = motion(-1.045*np.ones(n1), x1_b, p1_b, p2_b, n1, 1)
    plt.plot(t_short[t_short <= 2.5], x_b[t_short <= 2.5], color='blue')
    
    
    # Setting moto bello 
    print('setting up')  #############################################
    # Creo x
    x = init_cond(array_in=-1.047*np.ones(m), init=True, mode='ones')
    # Creo x1
    x1 = init_cond(init=False, mode='zeros')
    # Creo p1:
    p1 = init_cond() 
    # Creo p2
    p11,p12 = 0,0.105
    space = (p12-p11)/(m)
    p2 = init_cond(array_in=np.arange(p11, p12, space), init=True, mode='zeros') 
    print('End setting') #############################################
    
    # Motion
    x, p, s = motion(x, x1, p1, p2, n, m)
    
    # ricerca del cammino con l'azione più bassa.
    t = np.arange(0,n*dt,dt) + 2.5
    p2 = p**2/2 # Vettore con tutti gli incrementi delle azioni
    S = [j.sum() for j in p2.T] # Calcolo tutte le azioni come somme degli incrementi
    S_min_index = S.index(min(S[1:]),1) # trovo l'indice con la minima azione
    
    for i in range(1,m):
        if i == S_min_index: # Se ho il cammino con azione minima plotto linea più massiccia
            plt.plot(t, x[:,i], alpha=1,linewidth=1.1)
        else: 
            plt.plot(t, x[:,i], alpha=0.7,linewidth=0.9)
    plt.ylim(-1.2,0.5)
    plt.xlim(0, 12.5)
    plt.xlabel('t')
    plt.ylabel('x')
    #plt.savefig('../../figures/lez_12_strange_motion.png', dpi = 300)
    plt.show()
    
    # Plotto andamento della azione al variare del cammino
    plt.plot(np.arange(len(S)-1), S[1:]) # Plotto un grafico delle azioni per vedere se c'è un minimo
    plt.xlabel('Cammino')
    plt.ylabel('S')
    #plt.savefig('../../figures/lez_12_strange_motion_action.png', dpi = 300)
    plt.show()
    
main()
```
