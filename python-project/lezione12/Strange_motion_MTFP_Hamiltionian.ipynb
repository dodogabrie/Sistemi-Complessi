```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from scipy.stats import kde
```

```python
def get_param():
    dt = 0.01
    A = 0.3
    w = 1
    n = 1000
    m = 100
    return dt, A, w, n, m
```

```python
def f(x):
    return x - x**3

def step(x1, x, p2):
    p = get_param()
    dt, A = p[:2]
    return (f(x) + A * np.sin(x1) + p2) * dt

def step_x1():
    p = get_param()
    dt, w = p[0], p[2]
    return w * dt

def step_p1(x1, p2):
    p = get_param()
    dt, A = p[:2]
    return (- A * np.cos(x1) * p2) * dt

def step_p2(x, p2):
    p = get_param()
    dt = p[0]
    return ( ( -1 + 3 * x**2 ) * p2 ) * dt
```

```python
def init_cond(array_in = 0, init=False, mode = 'zeros'):
    par = get_param()
    n, m = par[3], par[4]
    if init:
        if mode == 'ones':
            x_init = array_in                  # Condizioni iniziali
            x_aux = np.ones((n-1 , m))          # Resto della matrice
            fin = np.vstack((x_init,x_aux))  # Unisco i vettori
            return fin
        if mode == 'zeros':
            x_init = array_in                  # Condizioni iniziali
            x_aux = np.zeros((n-1 , m))          # Resto della matrice
            fin = np.vstack((x_init,x_aux))  # Unisco i vettori
            return fin
    else:
        if mode == 'zeros': 
            return np.zeros((n , m))
        if mode == 'ones':
            return np.ones((n , m))
```

```python
def motion(x, x1, p1, p2, n, m):
    '''Gestione del movimento del camminatore'''
    aux = x
    p = get_param()
    dt, A, w = p[:3]
    print('start cycling')
    s = []
    if m > 1: 
        for j in range(m):
            for i in range(n-1):
                if np.abs(x[i,j]) <= 1.5:
                    x1[i+1,j] = x1[i,j] + step_x1()
                    x[i+1,j]  = x[i,j]  + step( x1[i,j], x[i,j], p2[i,j] ) 
                    p1[i+1,j] = p1[i,j] + step_p1( x1[i,j], p2[i,j] )
                    p2[i+1,j] = p2[i,j] + step_p2( x[i,j], p2[i,j] )
                else:
                    s.append(i)
                    x[i+1,j] = x[i,j]
    else:
        for j in range(m):
            for i in range(n-1):
                if np.abs(x[i]) <= 1.5:
                    x1[i+1] = x1[i] + step_x1()
                    x[i+1]  = x[i]  + step( x1[i], x[i], p2[i] ) 
                    p1[i+1] = p1[i] + step_p1( x1[i], p2[i] )
                    p2[i+1] = p2[i] + step_p2( x[i], p2[i] )
                else:
                    s.append(i)
                    x[i+1] = x[i]
    return x, s
```

```python
def main():
    par = get_param()
    n, m = par[3], par[4] # prendo i parametri del problema
    dt, w = par[0], par[2]
    #plt.figure(figsize=(20,10))
    
    # Setting assi e limiti del moto
    T = 2*np.pi/w
    n1 = 2*n
    t = np.arange(-T,n1*dt-T,dt) + 2.5
    x_b = -1.045*np.ones(n1)
    x1_b = np.zeros(n1)
    p1_b = np.zeros(n1)
    p2_b = np.zeros(n1)
    x_b, s = motion(x_b, x1_b, p1_b, p2_b, n1, 1)
    plt.plot(t, x_b, color = 'black', linestyle='--')
    
    x_b, s = motion(.945*np.ones(n1), x1_b, p1_b, p2_b, n1, 1)
    plt.plot(t, x_b-1, color='black', linestyle='--')
    
    t_short = np.arange(-T, n1*dt - T,dt) + 2.5 
    x_b, s = motion(-1.045*np.ones(n1), x1_b, p1_b, p2_b, n1, 1)
    plt.plot(t_short[t_short <= 2.5], x_b[t_short <= 2.5], color='blue')
    
    
    # Setting moto bello 
    print('setting up')  #############################################
    # Creo x
    x = init_cond(array_in=-1.047*np.ones(m), init=True, mode='ones')
    # Creo x1
    x1 = init_cond(init=False, mode='zeros')
    # Creo p1:
    p1 = init_cond() 
    # Creo p2
    p11,p12 = 0,0.105
    space = (p12-p11)/(m)
    p2 = init_cond(array_in=np.arange(p11, p12, space), init=True, mode='zeros') 
    print('End setting') #############################################
    
    # Motion
    x, s = motion(x, x1, p1, p2, n, m)
    
    t = np.arange(0,n*dt,dt) + 2.5
    for i in range(1,m):
        plt.plot(t, x[:,i], alpha=0.7,linewidth=0.9)
    
    plt.ylim(-1.2,0.5)
    plt.xlim(0, 12.5)
    plt.xlabel('t')
    plt.ylabel('x')
    plt.savefig('../../figures/lez_12_strange_motion.png', dpi = 300)
    plt.show()
main()
```
