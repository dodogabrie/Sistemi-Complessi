```python
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve
import warnings
warnings.filterwarnings("error")
import logging
import sys
sys.path.insert(0,'../fortran-package') # carico la directory con i pacchetti
from maps import standardmap
```

```python
def main():
    plt.figure(figsize=(6, 6))
    s = 5000
    k = 0.97
    l1 = np.arange(0., .9, 0.06) # Range di parametri iniziali per x
    l2 = np.arange(0., .9, 0.06) # Range di parametri iniziali per y
    colors = iter(cm.inferno(np.linspace(0, 1, len(l1)*len(l2)))) # colormap
    print(len(l1)*len(l2)*s)
    for i in l1: 
        for j in l2:
            x = np.ones(s)*i
            y = np.ones(s)*j
            standardmap(x, y, k) # richiamo la subroutine di fortran standardmao (in questa cartella)
            x[x<0] = x[x<0]+1
            y[y<0] = y[y<0]+1
            plt.scatter(x, y, s = 0.01, alpha=0.5, color = next(colors))
    plt.xlim(0., 1.)
    plt.ylim(0., 1.)
    plt.title('Mappa standard')
    plt.xlabel('q')
    plt.ylabel('p')
    #plt.savefig('../../figures/19_standard_map.png', dpi = 300)
    plt.show()
main()
```
