```python
%matplotlib inline
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve
import warnings
warnings.filterwarnings("error")
import logging
import sys
sys.path.insert(0,'../fortran-package') # carico la directory con i pacchetti
import maps
from matplotlib import animation
from IPython.display import HTML
```

```python
def logistic(xx, xy, n, lamda):
    for i in range(n-1):
        if i%2 == 1: # Se  uno step dispari (linee orizzontali)
            xx[i+1] = xy[i] # la x deve andare dalla vecchia y
            xy[i+1] = xy[i] # la y non deve cambiare
        else: # se  uno step pari (linee verticali)
            xy[i+1] = 4 * lamda * xx[i] * ( 1 - xx[i]) # Scrivo la mappa
            xx[i+1] = xx[i] # la x non deve cambiare
    return xx, xy

def main():
    """
    plotto i primi step della mappa logistica
    """
    lamda = 3/4 # valore di lambda selezionato per questa mappa
    row = 4 # numero di righe del plot
    col = 3 # numero di colonne del plot
    fig, ax = plt.subplots(row, col, figsize=(6, 8))
    x = np.linspace(0, 1, 100) # inizializzo un array per fare parabola e retta
    y = x
    par = 4 * lamda * x * ( 1 - x)
    k = 2
    for i in range(row):
        for j in range(col): 
            xx = np.ones(k)*0.2 # condizioni iniziali fisse ogni volta
            xy = np.zeros(k) # array ausiliario (vedi la funzione sopra)
            ax[i][j].plot(x, y, linewidth=0.5) # plotto retta
            ax[i][j].plot(x, par, linewidth=0.5) # plotto parabola
            
            xx, xy = logistic(xx, xy, k, lamda) # faccio la mappa
            ax[i][j].plot(xx, xy) 
            ax[i][j].set_xlim(0,1)
            ax[i][j].set_ylim(0,1)
            if j > 0:
                ax[i][j].get_yaxis().set_visible(False)
            else:
                ax[i][j].set_ylabel(r'$x_{n+1}$')
            if i < row-1:
                ax[i][j].get_xaxis().set_visible(False)
            else:
                ax[i][j].set_xlabel(r'$x_{n}$')
            ax[i][j].set_title(f'Step {k-1}')
            k += 1
    plt.tight_layout()
    #plt.savefig('../../figures/20_mappa_logistica_0.75.png', dpi = 200)
    plt.show()
main()
```

```python
def lamda_plot():
    n = 20000 # Attenzione!!! Non esagerare....
    m = 1000 # Attenzione che esplode il pc....
    pp = 60 # Plotto gli ultimi 60 pt per ogni traiettoria
    fig = plt.figure(figsize=(8,7))
    lamdas = np.linspace(0.7, 1, m) # selezione di lamda 
    xx = np.ones(n)*0.5 # parametro iniziale in array
    yy = xx # array ausiliario per velocizzare
    fin_x = [] # ausiliario per velocizzare
    fin_y = [] # \\
    print(f'plotterai {pp*m} points: overflow?')
    lyap = []
    for i,l in enumerate(lamdas):
        ### Belluire######
        cage = np.floor(10*(i+1)/m)
        perc_string = '|' + chr(9608)*int(cage) + '#'*int(10-cage) + '| '
        print(perc_string, end="\r")
        ##################
        x = np.ones(pp-1)*l # creo array verticale per le x
        fin_x.append(x) # lo aggiungo a questo array 
        yy = maps.logisticmap(xx, l) # faccio la mappa logistica
        fin_y.append(yy[n-pp:n-1]) # aggiungo agli array solo gli ultimi elementi termalizzati
        lyap.append(1/len(yy) * np.sum(np.log(np.abs((4 * l * ( 1 - 2 *  yy[1:]))))) )
    print('Se vedi la barra su pi linee stai inconsciamente facendo calcolo parallelo...')
    fin_x = np.reshape(np.array(fin_x), (pp-1)*m) # Faccio un reshape per avere array 1D
    fin_y = np.reshape(np.array(fin_y), (pp-1)*m) # Faccio reshape per array 1D
    plt.scatter(fin_x, fin_y, s = 0.03, c='black', alpha=0.3)
    plt.plot(lamdas, lyap, c='red')
    plt.plot(lamdas, np.zeros(len(lamdas)), linestyle='--', linewidth=0.5, c='black')
    plt.ylim(-0.5,1)
    plt.xlim(0.7,1)
    plt.ylabel(r'$\left<\sigma\right>$')
    plt.xlabel(r'$\lambda$')
    plt.savefig('../../figures/20_mappa_logistica_lyapunov.png', dpi = 300)
    plt.show() 
lamda_plot()
```
