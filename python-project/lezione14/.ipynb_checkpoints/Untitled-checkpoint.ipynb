```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def phase_plot():
    g,L = 1,1
    xvalues, yvalues = np.meshgrid(np.arange(-6, 6, 0.005), np.arange(-3, 3, 0.005))
    print('Meshgrid creata')
    xdot = yvalues
    ydot = -g/L*np.sin(xvalues)
    print('Creo il plot...')
    plt.streamplot(xvalues, yvalues, xdot, ydot, color='black', linewidth=0.5)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel('$x$')
    plt.ylabel('$v$')
    plt.show()
phase_plot()
```
