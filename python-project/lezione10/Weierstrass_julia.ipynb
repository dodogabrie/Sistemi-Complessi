## Importo Pacchetti

```julia
using Random, Distributions, Plots
```

## Definizione distribuzione di probabilità:

```julia
function P(x, M::Int64=30, b::Int64=2, a::Int64=1, N::Int64=10)
    """Distribuzione di probabilità per Random Walk di Weierstrass""" 
    norm = (M-1)/(2M)
    sum = 0
    for j = 0:N
        if x == b^j*a || x ==-b^j*a
        sum = sum + 1/M^j
        end
    end
    return sum
end

function MH(delta::Int64=1000, N::Int64=50000, b_in::Int64=2, M_in::Int64=30)
    """Algoritmo Metropolis Improvvisato"""
    i = 0
    x = [1.]
    x_s = [0.]
    naccept = 0
    counter = 0
    while naccept < N
        counter = counter + 1
        xt = last(x) + floor(delta * (2 * rand(Uniform(0, 1)) - 1))
        w = P(xt, M_in, b_in)/P(last(x), M_in, b_in)
        r = rand(Uniform(0, 1))
        if w > r
            append!( x_s, last(x_s) + xt )
            append!( x, xt )
            naccept = naccept + 1
        end
        i = i+1
    end
    println("#iterations = $i")
    println("naccept/counter = $(naccept/counter)")
    #h = histogram(x, bins = 600)
    #display(h)
    return x_s[Int64(N/2):end] .- x_s[Int64(N/2)]
end
```

## Applico il metropolis su x e y

```julia
delta = 1000
N = 50000
b = 2
M = 30

x = MH(delta, N, b, M)
y = MH(delta, N, b, M)
# @time MH(delta, N, b, M) --> 5 secondi per loop con più di 2*10^6 iterazioni
```
