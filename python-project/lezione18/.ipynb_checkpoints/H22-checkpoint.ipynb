```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import warnings
warnings.filterwarnings("error")
```

```python
def motion_pq(val, t):
    """
    Equazioni di Hamilton nello spazio p, q (non utilizzate nel codice)
    """
    x, y, px, py = val
    diff = [px * ( 1 - ( px**2 + x**2 ) - py**2),
           py * ( 4 * py**2 + 3 * y**2 ),
           - x * (1 - ( px**2 + x**2 ) - py**2 ),
           - y * ( 3 * py**2 + 2 * y**2)]
    return diff
```

```python
def rev_canonical( J, phi ):
    """Faccio la trasformazione canonica per tornare nelle p,q"""
    p = - np.sqrt( 2*J ) * np.sin( phi )
    q = np.sqrt( 2*J ) * np.cos( phi )
    return p,q

def my_integrator(t, alfa, init):
    """Definisco un Eulero al secondo ordine"""
    n = len(t)
    dt = t[1]-t[0]
    J1 = init[0]*np.ones(n) 
    J2 = init[1]*np.ones(n)
    phi1 = init[2]*np.ones(n)
    phi2 = init[3]*np.ones(n)
    x = np.array([J1,J2,phi1,phi2])
    for i in range(n-1):
        x_aux = x[:,i] + np.array(motion(t, x[:,i], alfa))*dt
        x[:,i+1] = 1/2*(x_aux + x[:,i] + np.array(motion(t, x_aux, alfa))*dt)
    return x

def motion(t, val, alfa):
    """
    Questa funzione definisce gli incrementi secondo il formato utile a 'solve_ivp' di scykit_learn.
    Se si vuole provare a fine codice c' uno script commentato, pu essere sostituito a my_integrator per vedere
    cosa cambia...
    """
    J1, J2, phi1, phi2 = val
    diff = [0,
            2 * alfa * J2 * ( J1 - J2 ) * np.sin( 2 * phi2 ),
            1 - 2 * J1 - J2 + alfa * J2 * np.cos( 2 * phi2 ) ,
            - J1 + 6 * J2 + alfa * ( J1 - 2 * J2 ) * np.cos( 2 * phi2 )]
    return diff

def energy(J1, J2, phi1, phi2, alfa):
    """L'hamiltoninana: per valutare l'energia"""
    return J1 - J1**2 - J1*J2 + 3 * J2**2 + alfa * J2 * ( J1 - J2 ) * np.cos( 2 * phi2 )

def eval_J2(J1, phi, E, alfa):
    """
    Valutazione delle condizioni iniziali su J12:
    stiamo cercando di variare le costanti iniziali lasciando fissa l'energia, per far questo una delle 4 
    variabili  fissata dalla Hamiltoniana.
    Invertendo l'Hamiltoninana si ottiene J2 in funzione di tutto il resto.
    """
    a = 3 - alfa * np.cos( 2 * phi )
    b = -J1 + alfa * J1 * np.cos( 2 * phi )
    c = J1 - J1**2 - E
    return (-b + np.sqrt(b**2 - 4 * a * c) )/(2 * a), (-b - np.sqrt(b**2 - 4 * a * c) )/(2 * a)

def eval_phi2(J1, E, alfa):
    """
    Ahim non  sempre possibile trovare un J2 reale, a seconda dell'angolo di partenza il J2 iniziale
    potrebbe essere complesso.
    Questa funzione assicura che, nei casi di interesse, si otterr un J2 reale (ovvero il main non dar errore).
    """
    a0 = 4*J1**2
    b0 = J1-J1**2-E
    a = a0
    b = 3*b0-2*a0
    c = -3*b0
    x_p = (- b + np.sqrt(b**2 - 4 * a *c))/(2 * a)
    x_m = (- b - np.sqrt(b**2 - 4 * a *c))/(2 * a)
    phi_m = np.arccos(x_m/alfa)/2
    return phi_m


def solver(t, init, alfa):
    """
    Questa funzione calcola l'intera evoluzione temporale delle J,phi facendo l'integrazione con my_integrator (qui pu essere sostituito il solver 
    commentato in fondo), inoltre fa la trasformazione canonica per ottenere py, y.
    """
    J1_in, J2_in, phi1_in, phi2_in = init
    sol = my_integrator(t, alfa, init)
    J1, J2, phi1, phi2 = sol # (come scrivere sol[0,:], sol[1,:], sol[2,:], sol[3,:])
    p, y = rev_canonical(J2, phi2)
    return p, y

def main():
    """
    Funzione main: fa da collante per tutte le funzioni sopra e inoltre plotta i risultati.
    """
    plt.figure(figsize=(7,7))
    # Energia e parametro della perturbazione
    alfa = 0.1
    E = 0.18
    
    ris = 0.243359092 # Valore atteso di J1 per la risonanza (in realt trovato a tentativi...
    
    j_in = [0.24, ris, ris+0.0005, ris+0.0021] # Array di valori iniziali per J1

    for J1_in in j_in: # al variare di J1 calcolo tutto e plotto
        print(J1_in)
        try: # Se da errore si ferma: sicuramente l'errore  un J1 senza soluzioni
            t = np.linspace(0, 401, 5001) # Intervallo temporale valutato
            
            # Valuto le condizioni iniziali
            phi1_in = 0 # useless
            phi2_in = eval_phi2(J1_in, E, alfa) 
            J2_plus, J2_minus = eval_J2(J1_in, phi2_in, E, alfa)
            
            # Plotto quelle con l'angolo positivo 
            init_plus = [J1_in, J2_plus, phi1_in, phi2_in]
            p_plus, y_plus = solver(t, init_plus, alfa)
            
            # Plotto quelle con l'angolo opposto ( definito da un arcos...)
            init_minus = [J1_in, J2_minus, phi1_in, -phi2_in]
            p_minus, y_minus = solver(t, init_minus, alfa)
            
            # Concateno e plotto i risultati
            y = np.concatenate( (y_plus, y_minus) )
            py = np.concatenate( (p_plus, p_minus) )
            plt.scatter(y, py, label=f'J1 = {J1_in:.4f}', s = 0.02, alpha=0.8)
        except RuntimeWarning: # Se avviene il runtimeWarning non fa nulla e stampa "errore" (passando al J successivo).
            print('Errore\n')

    lgnd = plt.legend()
    for handle in lgnd.legendHandles:
        handle.set_sizes([20.0])
    plt.grid(alpha=0.3)
    plt.title(fr'E={E}, $\alpha$={alfa}')
    plt.xlabel('y')
    plt.ylabel(r'$p_y$')
    #plt.savefig('../../figures/18_H22.png', dpi = 300)
    plt.show()
main()
```
