```python
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve
import warnings
warnings.filterwarnings("error")
import logging
import sys
sys.path.insert(0,'../fortran-package') # carico la directory con i pacchetti
from maps import standardmap
```

```python
def plot_map(l1, l2, s, k, ax):
    colors = iter(cm.inferno(np.linspace(0, 1, len(l1)*len(l2)))) # colormap
    for i in l1: 
        for j in l2:
            x = np.ones(s)*i
            y = np.ones(s)*j
            standardmap(x, y, k) # richiamo la subroutine di fortran standardmao (in questa cartella)
            x[x<0] = x[x<0]+1
            y[y<0] = y[y<0]+1
            ax.scatter(x, y, s = 0.01, alpha=0.5, color = next(colors))
    return ax

def main():
    fig, ax = plt.subplots( 2, 3, figsize=(8, 6))
    s = 5000
    k_vec = np.array([1/16, 1/14, 1/12, 1/10, 1/8, 1/7]) * 2 * np.pi
    k_string = ["1/16", "1/14", "1/12", "1/10", "1/8", "1/7"]

    low_x, up_x = 0., 0.2
    low_y, up_y = 0., 0.2
    l1_low = np.arange(low_x, up_x, 0.02) # Range di parametri iniziali per x
    l2_low = np.arange(low_y, up_y, 0.02) # Range di parametri iniziali per y
    #colors = iter(cm.inferno(np.linspace(0, 1, len(l1)*len(l2)))) # colormap
    t, m = 0, 0
    for b in range(6):
        ax[t][m] = plot_map(l1_low, l2_low, s, k_vec[b], ax[t][m])
        ax[t][m].set_xlim(low_x, up_x)
        ax[t][m].set_ylim(low_y, up_y)
        if t == 1:
            ax[t][m].set_xlabel('q')
            ax[t][m].set_xticks([0., 0.05, 0.1, 0.15])
        else:
            ax[t][m].get_xaxis().set_visible(False)
        if m == 0:
            ax[t][m].set_ylabel('p')
            ax[t][m].set_yticks([0., 0.1])
            #ax[t][m].set_yticklabels(['0', '0.05', 0.1, 0.15])
        else:
            ax[t][m].get_yaxis().set_visible(False)
        ax[t][m].set_title('k = ' + k_string[b])
        if m ==2:
            m = 0
            t = 1
        else:
            m += 1

    fig.suptitle('Mappa standard')
    plt.subplots_adjust(wspace=0.1, hspace=0.15)
    #plt.savefig('../../figures/21_standard_maps.png', dpi = 300)
    plt.show()
main()
```
