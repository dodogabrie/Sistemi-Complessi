```python
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve
import warnings
warnings.filterwarnings("error")
import logging
from henonmap import henonmap
```

```python
def main():
    plt.figure(figsize=(6,6))
    s = 5000
    k = 0

    alfa = np.arccos(0.24) # alfa  il parametro pi delicato del problema, occhio alla sua scelta...
    x_in = [-0.3199] # Guessing del punto fisso
    y_in = [-0.3199] # //
    l1 = np.arange(-0.9, 0.9, 0.15) # Range di parametri iniziali per x
    l2 = np.arange(-0.9, 0.9, 0.15) # Range di parametri iniziali per y
    colors = iter(cm.inferno(np.linspace(0, 1, len(l1)*len(l2)))) # colormap
    print(len(l1)*len(l2)*s)
    for i in l1: 
        for j in l2:
            x = np.ones(s)*i
            y = np.ones(s)*j
            henonmap(x, y, alfa) # richiamo la subroutine di fortran henonmap (in questa cartella)
            xt = x[x<1] # Evito di plottare dati molto lontani dall'origine (anche 1e100..)
            yt = y[x<1] # idem per y
            plt.scatter(xt, yt, s = 0.01, alpha=0.5, color = next(colors))
            
    for i, j in zip(x_in, y_in): # Parametri iniziali pi interessanti li plotto con marker pi spessi
        x = np.ones(s)*i
        y = np.ones(s)*j
        henonmap(x, y, alfa)
        xt = x[x<1]
        yt = y[x<1]
        plt.scatter(xt, yt, s = 0.3, color='black')
    print(f'lavori con un angolo alfa={alfa}')
    plt.xlim(-0.8, 0.9)
    plt.ylim(-0.8, 0.9)
    plt.title('Mappa di Henon')
    plt.xlabel('x')
    plt.ylabel('y')
    #plt.savefig('../../figures/18_Henon_map.png', dpi = 300)
    plt.show()
main()
```
